```python
#WS cuts
S1_min=0
S1_max=20
S2raw_min=450
```

<h1> Sweep g1 and find signal acceptance </h1>

```python
##############################################################################
#Generate Flat ER and 50 GeV WIMP NR spectra:
##############################################################################
#LZ
NEST =NEST_setup(Det='LZ', nFold=3, g1=0.075, e_lifetime=850, eff_extract=0.95, custom=True)
```

```python
Er, Nph, Ne, S1, S2, S1c, S2c, WmpRate, Det_exposure_factor = WIMP2NphNe(NEST=NEST, mWmp=50,nSim=1e5,showFig=True)
```

```python
g1s=np.arange(0.05,0.45,0.05)

start = 0.0
stop = 1.0
number_of_lines= len(g1s)
cm_subsection = linspace(start, stop, number_of_lines) 
colors = [ cm.jet(x) for x in cm_subsection ]

for ii, g1 in enumerate(g1s):
    Er, Nph, Ne, S1, S2, S1c, S2c, WmpRate, Det_exposure_factor = \
    WIMP2NphNe(NEST=NEST_setup(Det='LZ', g1=g1, custom=True), mWmp=50,nSim=1e5,showFig=False)
    WS_cut=(S1c>0) & (S2>S2raw_min)
    bins=np.arange(0,50,1)
    n, edges= histogram(S1c[WS_cut],bins=bins)
    accept=sum(WS_cut)/len(Er) #ratio of events passing threshold
    plt.step(edges[1:],n, linewidth=2, color=colors[ii], label='g1='+str(g1)+', sig accpt={:0.2f}'.format(accept))

xlabel('S1c [phd]')
ylabel('relative rate')
legend(loc='upper right', fontsize=20)
```

<h1> Get integral count vs. g1 for a 50 GeV WIMP </h1>

```python
pylab.rcParams['figure.figsize'] = (11.0, 8.0) # set size of figures"
plt.rcParams.update({'font.size': 24})

start = 0.0
stop = 1.0
number_of_lines= 9
cm_subsection = linspace(start, stop, number_of_lines) 
colors = [ cm.jet(x) for x in cm_subsection ]

g1s=np.arange(0.05,0.40,0.05)
WIMPs=[10, 15, 20, 25, 30, 40,50,100,1000]
nSim=1e5
count_vs_wimp=[]

for jj, mWmp in enumerate(WIMPs):
    count = []
    for ii, g1 in enumerate(g1s):
        Er, Nph, Ne, S1, S2, S1c, S2c, WmpRate, Det_exposure_factor = \
        WIMP2NphNe(NEST=NEST_setup(Det='LZ', g1=g1, custom=True), mWmp=mWmp,nSim=nSim,showFig=False)
        count.append(size(S1c[(S1c>0) & (S2>S2raw_min)])/nSim)
    
    count_vs_wimp.append(count)    
    plot(g1s,count, linewidth=3, label=str(mWmp)+' GeV', color=colors[jj])

grid('on')
xlabel('g1 [phd]')
ylabel('relative signal count')
legend(loc='lower right', fontsize=20)
```
