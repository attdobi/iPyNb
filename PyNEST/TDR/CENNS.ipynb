<h1> Plot 8B dependence vs. detector parameters </h1>

```python
import numpy as np
%pylab inline
pylab.rcParams['figure.figsize'] = (10.0, 8.0) # set size of figures"
plt.rcParams.update({'font.size': 24})
```

```python
Nfold=[2,3,4,5]
N_ATM=[0.514, 0.444, 0.382, 0.34]
N_B8=[55, 7.1, 0.556, 0.041]
N_HEP=[1.1, 0.233, 0.039, 0.006]
N_DSN=[0.14, 0.092, 0.061, 0.043]

g1s=[0.05,0.075, 0.12]
g1_ATM=[0.434, 0.444, 0.411]
g1_B8=[0.85, 7.1, 37.046]
g1_HEP=[0.051, 0.233, 0.846]
g1_DSN=[0.06, 0.092, 0.135]
```

```python
#Plot N fold coincidence dependence
plt.plot(Nfold, N_B8,'--or',markersize=10, label='B8')
plt.plot(Nfold, N_ATM,'--^b',markersize=10, label='ATM')
plt.plot(Nfold, N_HEP,'--sk',markersize=10,label='HEP')
plt.plot(Nfold, N_DSN,'--*g',markersize=10,label='DSN')
yscale('log')
ylabel('Count')
xticks([2,3,4,5])
xlabel('N-fold coincidence')
xlim([1.5,5.5])
legend(numpoints=1)
#grid('on')
```

```python
#Plot g1 dependence
plt.plot(g1s, g1_B8,'-or',markersize=10, label='B8')
plt.plot(g1s, g1_ATM,'-^b',markersize=10, label='ATM')
plt.plot(g1s, g1_HEP,'-sk',markersize=10,label='HEP')
plt.plot(g1s, g1_DSN,'-*g',markersize=10,label='DSN')
yscale('log')
ylabel('Count')
xticks(np.arange(0.04,0.13,0.02))
xlabel('light collection fraction')
xlim([0.04,0.13])
legend(numpoints=1,fontsize=20,loc='upper left')
#grid('on')
```
