```python
from __future__ import division
import sys
home_dir=!echo "$HOME"
sys.path.insert(2,home_dir[0]+'/iPyNb/NERSC_Import') 

from pyROOT_import import *
from mpl_numpy_scipy_import import *
from ROOT import TEllipse
```

```python
path='/project/projectdirs/lux/data/golden/Makrus_filter/Run03_1S1_10S2.root'
```

```python
chain = TChain('filter')
chain.Add(path)
```

```python
g1=0.1167
g2=12.05
```

```python
c1.Clear()
c1 = rootnotes.default_canvas(name='icanvas1')
```

```python
fid_vol=TCut('S2.fDrift>35 && S2.fDrift<308 && S2.fRad_cor<22')
exclusion_cut=TCut('S1.fPartner==0 && excluded==0 ')
area_cut=TCut('S1.fArea>1 && S1.fArea_raw<10**7 && S2.fArea_raw>150 && S2.fArea>100 && S2.fArea<10**6')
bad_area_cut=TCut('((goodArea<=630 && badArea < 80) || (goodArea>630 && badArea < 80+0.095*(goodArea-630)))')
combined_cut=TCut(fid_vol.GetTitle() + '&&'+ area_cut.GetTitle() + '&&' + bad_area_cut.GetTitle() );


chain.SetMarkerStyle(7)
chain.SetMarkerSize(10)

chain.Draw('log10(S1.fArea_raw)',combined_cut)


c1
```
