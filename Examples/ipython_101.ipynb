<h3> a cooler plot</h3> <br/> <h3 style="color:purple;"> Surface plot with mesh grid</h3>

```python
import scipy.stats as st
```

```python
phi_m=linspace(0,2*pi,100) #setting up a linear space from 0 to 2pi with 100 elements. note: logspace does the same thing... in log space "mind blown"
phi_p=linspace(0,2*pi,100)
X,Y = meshgrid(phi_p,phi_m)

#Defining some blah blah blah function here :) ... to make a cool looking plot
alpha = 0.7
phi_ext = 2 * pylab.pi * 0.5
def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * pylab.cos(phi_p)*pylab.cos(phi_m) - alpha * pylab.cos(phi_ext - 2*phi_p)
#/end

Z = flux_qubit_potential(X, Y).T

p=pcolor(X/(2*pi),Y/(2*pi),Z,cmap=cm.RdBu)
cb = colorbar(p, shrink=0.5)
plt.title('A weird 2D potential field in KAUST colors')
plt.ylabel('y',fontsize=18)
plt.xlabel('x')
cb.set_label('V(x,y)')
```

<h3 style="color:'red;"> density scatter plot, (colored by local density) </h3>

```python
#figure()
x=randn(1,10000)
y=randn(1,10000)
xy=vstack([x,y])
z= st.gaussian_kde(xy)(xy)
scatter(x,y,c=z,s=10,edgecolor='',alpha=0.7) #alpha sets the transparency of the points
#plt.show()
#yscale('log')
xlabel('X')
ylabel('Y')
xlim([-4, 4])
ylim([-4, 4])
text(.1,3,"woot woot")
```
